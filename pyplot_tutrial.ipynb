```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.plot(range(10))
plt.xlim(2,20)
plt.ylim(0,10)
plt.xticks([3,8])
plt.yticks([4,7])
plt.xlabel('measured')
plt.ylabel('calculated')
plt.title('measured vs calculated')
plt.legend(loc='upper center',ncol=3)
plt.grid(True)
plt.savefig('hh.png',dpi=150)
```

Line plot :- plot(x,y)

```python
with plt.style.context('seaborn-notebook'):
    x = np.linspace(0,20)
    y1 = np.sin(x)
    y2 = np.sin(x-np.pi)
    plt.figure()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x,
             y1,
             color = 'black',
             linestyle='-',
             linewidth=2,
             marker = 's',
             markersize=6,
             label='y1')
    plt.plot(x,
             y2,
             color = 'red',
             linestyle='--',
             linewidth=2,
             marker = '^',
             markersize=6,
             label='y2')
    plt.legend()
```

Vertical Bar Charts :- bar(left,height)

```python
years = np.arange(2012,2015)
values = [2,5,9]
plt.figure()
plt.bar(years,values)
plt.figure()
values1 = [2,5,9]
values2 = [7,6,3]
plt.bar(years-0.2,
        values1,
        color='blue',
        edgecolor='red',
        width=0.4,
        align='center',
        label='y1')
plt.bar(years+0.2,
       values2,
       color='orange',
       edgecolor='black',
       width=0.4,
       align='center',
       label='y2')
plt.xticks(years,[str(year) for year in years])
plt.legend()
```

Horizontal Bar Chartss :- barh(x,y)

```python
categories = ['A','B','C','D','E']
values = [5,3,8,9,4]
plt.figure()
plt.barh(categories,values)
plt.yticks(np.arange(len(categories)),
          ['Categories {}'.format(x) for x in categories])
plt.figure()
plt.barh(-np.arange(len(categories)),values)
plt.yticks(-np.arange(len(categories)),
          ['Categories {}'.format(x) for x in categories])
plt.figure()
plt.barh(-np.arange(len(categories)),
         values,
        color='cyan',
        edgecolor='green',
        height=0.6,
        align='center')
plt.yticks(-np.arange(len(categories)),
          ['Categories {}'.format(x) for x in categories])
```

to display the horizontal bars in sorted fashion

```python
import pandas as pd

categories = ['A','B','C','D','E']
values = [5,3,8,9,4]
category_values = pd.DataFrame({'categories':categories,
                                'values':values})
category_values.sort_values('values',ascending=False,inplace=True)
categories = category_values['categories'].values
values = category_values['values'].values
plt.figure()
plt.barh(-np.arange(len(categories)),values)
plt.yticks(-np.arange(len(categories)),
          ['Categories {}'.format(x) for x in categories])
```

pie charts :- pie(sizes)

```python
counts = [17,14]
plt.figure(figsize=(4,4))
plt.pie(counts)
plt.figure(figsize=(4,4))
plt.pie(counts,
       colors=['red','green'],
       labels=['Category A','Other Category'],
       startangle=90,
       autopct='%1.1f%%')
```

scatter plot :- scatter(x,y)

```python
x=range(20)
y = np.arange(50,70) + (np.random.random(20)*10.)
plt.figure()
plt.scatter(x,y)
plt.figure()
plt.scatter(x,
           y,
           c=['red','yellow','green'],
           s = 40,
           marker='s',
           edgecolor='none')
```

--------Plotting Distributions---------- 

Histograms :- hist(x)

```python
column_data = np.random.normal(42,3,1000)
plt.figure()
plt.hist(column_data,
        color='blue',
        bins=50,
        normed=True,
        edgecolor='red')
```

Box plots :- boxplot(x)

```python
x = [np.random.normal(52,3,1000),
    np.random.normal(42,7,1000)]
plt.figure()
plt.boxplot(x)
plt.figure()
plt.boxplot(x,
           labels=['A','B'],
           notch=True,
           bootstrap=10000)
```

Subplots 

```python
plt.figure()
for i in range(1,7):
    plt.subplot(3,2,i)
    plt.title(i)
    plt.xticks([])
    plt.yticks([])
plt.tight_layout()
```

```python
years = np.arange(2010,2016)
plt.figure(figsize=(10,10))
for cat_num in range(1,17):
    plt.subplot(4,4,cat_num)
    y_vals = np.arange(10,16) + (np.random.random(6)*cat_num/4.)
    plt.plot(years,y_vals)
    plt.ylim(10,18)
    plt.xticks(years,[str(year) for year in years])
    plt.title('Category{}'.format(cat_num))
plt.tight_layout()
```
